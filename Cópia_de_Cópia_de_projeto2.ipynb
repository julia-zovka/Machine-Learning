####**EXTRAÇÃO DOS DATAFRAMES**

Agora que todos os arquivos (benignos, mirai e gafgyt) estão descompactados e disponíveis como `.csv`, vamos carregá-los em um único DataFrame para cada dispositivo iot

```python
# ista com os 9 nomes
devices = [
    'Danmini_Doorbell',
    'Ecobee_Thermostat',
    'Ennio_Doorbell',
    'Philips_B120N10_Baby_Monitor',
    'Provision_PT_737E_Security_Camera',
    'Provision_PT_838_Security_Camera',
    'Samsung_SNH_1011_N_Webcam',
    'SimpleHome_XCS7_1002_WHT_Security_Camera',
    'SimpleHome_XCS7_1003_WHT_Security_Camera'
]
```

```python
output_folder = 'Consolidated_CSVs'
os.makedirs(output_folder, exist_ok=True)

for device_name in devices:
    search_path = f'{device_name}/*.csv'  # todos os CSVs do dispositivo
    device_files = glob.glob(search_path)

    if not device_files:
        print(f"Nenhum CSV encontrado para {device_name}. Pulando...")
        continue

    df_list = []
    for filepath in device_files:
        filename = os.path.basename(filepath)
        # Extrai o nome base do arquivo (sem extensão) e então usa o split para pegar o tipo de ataque
        attack_type = os.path.splitext(filename)[0] # Remove .csv
        attack_type = attack_type.split('_')[0] # Pega a primeira parte antes do '_'


        try:
            df_aux = pd.read_csv(filepath, low_memory=False)
            df_aux['Attack_Type'] = attack_type
            # df_aux['Source_File'] = source_name  # aqui fica o nome limpo do CSV
            df_list.append(df_aux)
        except Exception as e:
            print(f"ERRO ao ler {filepath}: {e}")

    # Concatena todos os CSVs do dispositivo
    if df_list:
        df_device = pd.concat(df_list, ignore_index=True)
        df_device.columns = df_device.columns.str.strip()

        # Salva o CSV consolidado do dispositivo
        output_path = os.path.join(output_folder, f"{device_name}.csv")
        df_device.to_csv(output_path, index=False)

        print(f"{device_name} consolidado em {output_path} com {len(df_device):,} linhas.")
```

1. ANÁLISE DE VARIÁVEIS NUMÉRICAS

```python
# Análise dos outliers de cada variável
for var in numeric_vars:
    if var in df.columns:
        print(f"\n📊 {var.upper()}:")
        data = df[var].dropna()


        # Outliers (IQR method)
        Q1 = data.quantile(0.25)
        Q3 = data.quantile(0.75)
        IQR = Q3 - Q1
        outliers = data[(data < Q1 - 1.5*IQR) | (data > Q3 + 1.5*IQR)]
        print(f"  Outliers: {len(outliers)} ({len(outliers)/len(data)*100:.1f}%)")
```

```python
# Visualizações das variáveis numéricas
total_vars = len(numeric_vars)
cols = 5
rows = (total_vars + cols - 1) // cols  # mesmo padrão, só calcula número de linhas

fig, axes = plt.subplots(rows, cols, figsize=(18, rows * 3))
fig.suptitle('Análise Univariada - Variáveis Numéricas', fontsize=16)

# Garantir que axes seja 2D sempre
axes = np.array(axes).reshape(rows, cols)

for i, var in enumerate(numeric_vars):
    if var in df.columns:
        row = i // 5
        col = i % 5
        # Histograma + Densidade
        axes[row, col].hist(df[var].dropna(), bins=30, alpha=0.7, density=True)
        df[var].dropna().plot.density(ax=axes[row, col], color='red')
        axes[row, col].set_title(f'{var} - Distribuição')
        axes[row, col].set_ylabel('Densidade')

# Remove subplots vazios
for j in range(i + 1, rows * cols):
    fig.delaxes(axes.flatten()[j])

plt.tight_layout()
plt.show()
```

```python
categorical_vars = df.select_dtypes(include=['object']).columns.tolist()
print(f"Variáveis categóricas: {categorical_vars}")
```

```python
# Análise de frequências
for var in categorical_vars:
    if var in df.columns:
        print(f"\n📊 {var.upper()}:")
        freq_table = df[var].value_counts()
        freq_percent = df[var].value_counts(normalize=True) * 100

        freq_df = pd.DataFrame({
            'Frequência': freq_table,
            'Percentual': freq_percent
        })
        print(freq_df.head(10))

        # Medida de concentração (Índice de Herfindahl)
        herfindahl = (freq_percent/100).pow(2).sum()
        print(f"  Índice de Herfindahl: {herfindahl:.3f}")
        print(f"  Entropia: {stats.entropy(freq_table):.3f}")
```

```python
# 1. Selecionar somente variáveis numéricas
numeric_df = df.select_dtypes(include=[np.number])

# 2. Calcular a matriz de correlação completa
corr_matrix = numeric_df.corr().abs()

# 3. Somar as correlações de cada variável para achar as mais "correlacionadas"
corr_scores = corr_matrix.sum().sort_values(ascending=False)

# 4. Selecionar as 20 principais variáveis
top20_vars = corr_scores.head(10).index

# 5. Criar a matriz de correlação apenas com essas 20
top20_corr_matrix = numeric_df[top20_vars].corr()
```

```python
# Identificar correlações fortes
strong_corr = []
for i in range(len(correlation_matrix.columns)):
    for j in range(i+1, len(correlation_matrix.columns)):
        corr_val = correlation_matrix.iloc[i, j]
        if abs(corr_val) > 0.5:  # Correlação forte
            strong_corr.append({
                'var1': correlation_matrix.columns[i],
                'var2': correlation_matrix.columns[j],
                'correlation': corr_val
            })
```

```python
if strong_corr:
    print("\n CORRELAÇÕES FORTES (|r| > 0.5):")
    for corr in strong_corr:
        print(f"  {corr['var1']} ↔ {corr['var2']}: {corr['correlation']:.3f}")
```

```python
# 2. ANÁLISE TARGET vs VARIÁVEIS NUMÉRICAS
fig, axes = plt.subplots(2, 2, figsize=(15, 12))
fig.suptitle(f'Análise Bivariada: {target_var} vs Variáveis Numéricas', fontsize=16)
numeric_for_analysis = ['Age', 'Fare', 'SibSp', 'Parch']
for i, var in enumerate(numeric_for_analysis):
    if var in df.columns:
        row = i // 2
        col = i % 2

        # Box plot por categoria do target
        sns.boxplot(data=df, x=target_var, y=var, ax=axes[row, col])
        axes[row, col].set_title(f'{var} por {target_var}')

        # Teste estatístico (t-test)
        group_0 = df[df[target_var] == 0][var].dropna()
        group_1 = df[df[target_var] == 1][var].dropna()

        if len(group_0) > 0 and len(group_1) > 0:
            t_stat, p_value = stats.ttest_ind(group_0, group_1)
            axes[row, col].text(0.05, 0.95, f'p-value: {p_value:.4f}',
                               transform=axes[row, col].transAxes,
                               bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))
plt.tight_layout()
plt.show()
```
