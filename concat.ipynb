```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file1 = "housing_price.csv"
```

```python
df_1= pd.read_csv(file1)
```

```python
df_encoded = pd.get_dummies(df_1, columns=["Neighborhood"])

print(df_encoded.head())
```

HISTOGRAMA

```python
plt.hist(df_1["Price"], bins=20, edgecolor="black")
plt.xlabel("Preço das casas")
plt.ylabel("Frequência")
plt.title("Histograma de preços das casas")
plt.show()
```

BOX PLOT

identificacao de anomalias

```python
plt.boxplot(df_1["Bedrooms"])
plt.title("Box plot do número de quartos")
plt.ylabel("Bedrooms")
plt.show()
```

```python
sns.boxplot(x="Neighborhood", y="SquareFeet", data=df_1)
plt.title("Box plot de Tamanho (SquareFeet) por Bairro")
plt.show()
```

CORRELAÇÃO

```python
sns.scatterplot(x="SquareFeet", y="Price", data=df_1)
plt.title("Preço vs. Tamanho da Casa")
plt.show()
```
